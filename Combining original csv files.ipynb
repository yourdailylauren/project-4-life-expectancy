```python
# Load the county data
county_data = pd.read_csv(file_path2)
```

```python
# Create a new column by appending 'county_full' with ', ' and 'state_id'
county_data['county_full_with_state'] = county_data['county_full'] + ', ' + county_data['state_id']
```

```python
# load the updated county data and the life expectancy data
life_expectancy_data = pd.read_csv(file_path1)
county_data_updated = pd.read_csv(file_path2)
```

```python
# Merge the two dataframes on the new 'county_full_with_state' column from county_data and 'County' column from life_expectancy_data
combined_data = pd.merge(life_expectancy_data, county_data_updated, left_on='County', right_on='county_full_with_state', how='left')
```

```python
# Split the 'Life Expectancy Range' into 'Life Expectancy Low' and 'Life Expectancy High'
combined_data[['Life Expectancy Low', 'Life Expectancy High']] = combined_data['Life Expectancy Range'].str.split('-', expand=True)
```

```python
# Convert the new columns to numeric type
combined_data['Life Expectancy Low'] = pd.to_numeric(combined_data['Life Expectancy Low'], errors='coerce')
combined_data['Life Expectancy High'] = pd.to_numeric(combined_data['Life Expectancy High'], errors='coerce')
```

```python
# Remove rows where 'Life Expectancy' column has NaN values
combined_data_cleaned = combined_data.dropna(subset=['Life Expectancy'])
```
